```python
import pandas as pd
from nba_api.stats.static import teams
from nba_api.stats.static import players
from nba_api.stats.endpoints import leaguegamefinder

import time
start_time = time.time()
```

```python
path = 'C:/Users/mitchtrainotti/Documents/NBA Project/Data/'
gameRecordName = '01gameRecord.xlsx'

gameRecordNameXls = '01gameRecord.xlsx'
```

```python
nba_teams = teams.get_teams()
# Select the dictionary for the Celtics, which contains their team ID
warriors = [team for team in nba_teams if team['abbreviation'] == 'GSW'][0]
warriors_id = warriors['id']
```

```python
nba_players = players.get_players()

warriors_players = []
steph = [player for player in nba_players if player['full_name'] == 'Stephen Curry']

warriors_players.append(steph)

warriors_players
```

```python
# Query for games where the Warriors were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=warriors_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games_2122 = games[games.SEASON_ID.str[-5:] == '22021']
games_2122 = games_2122[games_2122.GAME_ID.str[:4] == '0022']
games_2122['homeFlag'] = games_2122['MATCHUP'].str.contains('vs.')


games_2122.head()
```

```python
newPath = path + gameRecordName
writer = pd.ExcelWriter(newPath, engine = 'xlsxwriter')
games_2122.to_excel(writer, gameRecordName)
writer.save()
```

```python
newPath = path + gameRecordNameXls
writer = pd.ExcelWriter(newPath, engine = 'xlsxwriter')
games_2122.to_excel(writer, 'Sheet1')
writer.save()

print("--- %s seconds ---" % (time.time() - start_time))
```
